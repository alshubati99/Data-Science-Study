```python
iris = sns.load_dataset("iris")
print(iris.head(2))
```

```python
for name in names:
  cond = iris['species'] == name
  plt.scatter(x[cond],y[cond], label = name)
plt.legend()
plt.show()
```

```python
x_data = np.arange(0,150) # sequence of numbers in [0,149]
colors = {'setosa':'r', 'versicolor':'g', 'verginica':'b'}

# create figure and axis
fig , ax = plt.subplots()
# plot each
for name in names:
  cond = iris['species'] == name
  plt.scatter(x[cond],y[cond], color = colors[name], label = name)

# set title and labels
ax.set_title('Iris dataset')
ax.set_xlabel('sepal length')
ax.set_ylabel('sepal width')
ax.legend()
```

### Visualizing 2D functions

```python
def f(x,y):
  return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
x = np.linspace(0,5,50)
y= np.linspace(0,5,40)
X,Y = np.meshgrid(x,y)
Z= f(X,Y)
```

```python
plt.contour(X,Y,Z, color = 'k')
```

```python
plt.contour(X,Y,Z, 20, cmap = 'RdGy')
```

```python
# filled countour plots:
plt.contourf(X,Y,Z, 20, cmap = 'RdGy')
plt.colorbar()
```

### Contour Plots Using oo Interface

```python
# contour plots
delta = 0.025
x = np.arange(-3.0, 3.0, delta)#sequence of numbers from -3 to 3, difference = delta
y = np.arange(-2.0, 2.0, delta)

X,Y=np.meshgrid(x,y)
Z1 = np.exp(-X**2-Y**2)
Z2 = np.exp(-(X-1)**2-(Y-1)**2)
Z=(Z1*Z2)*2

fig, ax = plt.subplots(1,2)
fig.set_size_inches(14,7)

CS = ax[0].contour(X,Y,Z)
ax[0].clabel(CS, inline = True, fontsize=10)
ax[0].set_title('simplist default with labels')

CS = ax[1].contour(X,Y,Z, color = 'k') # negative contour default to dashed
ax[0].clabel(CS, inline = True, fontsize=11)
ax[0].set_title('single color - negative contour dashed')

```

### Grouping Bars

```python
labels = ['G1','G2','G3','G4','G5']
men_means = [20,34,30,35,27]
women_means = [25,32,34,20,25]

x = np.arange(len(labels)) # the label location
width = 0.35 # width of bar

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label = 'Men')
rects2 = ax.bar(x + width/2, women_means, width, label = 'Women')

# add some text for labels, title and custom x-axis tick labels
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x,labels)
ax.legend()

ax.bar_label(rects1, padding = 3)
ax.bar_label(rects2, padding = 33)
```

```python
labels = ['G1','G2','G3','G4','G5']
men_means = [20,34,30,35,27]
women_means = [25,32,34,20,25]

fig, ax = plt.subplots()
ax.bar(labels, men_means, label = 'Men')
ax.bar(labels, women_means,bottom= men_means ,label = 'Women')

# add some text for labels, title and custom x-axis tick labels
ax.set_ylabel('mean values')
ax.set_title('Scores by group and gender')
ax.set_xticks([0,1,2,3,4])
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

### Annotations

```python
# prepare dataset
x=[x for x in range(10)]
y=[5,2,6,3,6,3,6,9,2,8]
text = ['sfd','dfdf','sfd','xx','asd','xxv','asa','asd','asd','asd']

# plotting scatter plot
plt.scatter(x,y)
#loop for annotation of all points
for i in range(len(x)):
  plt.annotate(text[i], (x[i],y[i]+0.2))
# adjusting the scale of the axes
plt.xlim((-1,10))
plt.ylim((0,10))
plt.show()
```
