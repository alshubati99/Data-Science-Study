# Clustering Methods
- Clustering is the use of unsupervised techniques for goruping similar objects.
- Unsupervised = finding hidden structure within unlabeled data.
- Clustering answer the question: what are the natural occuring groupings?!
- Clustering Applications: image processing, medical and customer segmentation, biology in classifications of plants and animals and in human genetics.

- **K-Means Clustering:** <br>
  - Given a collection of objects, k-means identifies **k** clusters of objects based on th object's proximity to the center of the k groups. The center is determined as the arithmetic average(mean) of each cluster's n-D vector of attributes.

- **Steps** <br>
  1. Find k clusters from collection of M objects with n attributes.
  2. We will examine case of n=2 where object has 2 attributes.
  3. Since object has 2 attributes then it corresponds to point(x,y).
  4. The point that corresponds to cluster's mean is called *Centriod*.
  5. Now the k-means algorithm to find k clusters can be described in 3 steps.
    a. Choose value of k as initial guess (ex: 3)
    b. Compute the distance from each data point(x,y) to each centriod. Assign each point to closest centriod. (the distance can be expressed by Euclidean Distance Measure)
    c. Compute the centriod, center of mass for each newly defined cluster.
    `now the two steps above are repeated until the algorithm converges.`
  6. Convergence is reached when the computed centriods do not change or the centriods and assigned points oscillate back and forth from one iteration to the next. The latter case can occur when there are one or more points that are equal distances from the computed centriod.

- **K-Means Algorithm:** <br>
  1. Let D the set of samples to be clustered.
  2. Randomly choose number of clusters (k)
  3. Randomly initialize the centriods
  4. Classify each sample according to the rule that the point is closest to its centriod.
  5. Recompute the cluster centriod using the means.
  6. If there is any modifecations repeat from classifying each sample

- **Fuzzy k-means (c-means) Clustering Algorithm:** <br>
  1. In Fuzzy k-means clustering, each sample is allowed to belong to more than one cluster to some extend.
  2. same steps for k-means but this one computes the membership with a different formula

- **Measurements Units:** <br>
  1. The k-means algorithm identify different clusters depending on the choice of the units of measure.
  2. Widely used rescaling approach is to divide each attribute by the attribute's standard deviation and the resulting attributes will each have a standard deviation equal to 1 and will be without units.

- **Choosing Best k Value:** <br>
  1. Within Cluster Sum of Squares (WSS or ineria) = Total sum of distance squares between each sample and the corresponding centriod.
  2. The smaller the WSS value, the more coherent are the different clusters. When as many clusters are added as there are samples in the dataset, then the WSS value would be zero.
  3. Elbow-Method is used to find the optimal number of clusters using inertia (WSS) value.
  4. A heuristic using WSS metric can be examined to determine a reasonably optimal value of k.
  `Remember!: WSS is sum of squares of distances between each data point and its closest centroid`

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris['species'] = iris['species'].replace('setosa',0)
iris['species'] = iris['species'].replace('versicolor',1)
iris['species'] = iris['species'].replace('virginica',2)

X = iris.drop('species', axis = 1) #features matrix
X = X[['petal_length','petal_width']]
y = iris['species']
```

```python
clustering = KMeans(n_clusters = 2, random_state = 428)
clustering.fit(X)
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
colors = np.array(["Red","Green","Blue","Black","Cyan"])
plt.subplot(1,2,1)
plt.scatter(x=X["petal_length"], y=X["petal_width"], c=colors[y])
plt.title("Before KMeans Clustering")

plt.subplot(1,2,2)
plt.scatter(x=X["petal_length"], y=X["petal_width"], c=colors[clustering.labels_])
plt.title("KMeans Clustering")
```

When k = 3

```python
clustering = KMeans(n_clusters = 3, random_state = 428)
clustering.fit(X)
fig, axs = plt.subplots(1,2, figsize=(15,5))
colors = np.array(["Red","Green","Blue","Black","Cyan"])
plt.subplot(1,2,1)
plt.scatter(x=X["petal_length"], y=X["petal_width"], c=colors[y])
plt.title("Before KMeans Clustering")

plt.subplot(1,2,2)
plt.scatter(x=X["petal_length"], y=X["petal_width"], c=colors[clustering.labels_])
plt.title("KMeans Clustering")
```

```python
clustering = KMeans(n_clusters = 5, random_state = 428)
clustering.fit(X)
fig, axs = plt.subplots(1,2, figsize=(15,5))
colors = np.array(["Red","Green","Blue","Black","Cyan"])
plt.subplot(1,2,1)
plt.scatter(x=X["petal_length"], y=X["petal_width"], c=colors[y])
plt.title("Before KMeans Clustering")

plt.subplot(1,2,2)
plt.scatter(x=X["petal_length"], y=X["petal_width"], c=colors[clustering.labels_])
plt.title("KMeans Clustering")
```

Choosing Best K Value

```python
inertia_list = []
for num_clusters in range(1,10):
  clustering = KMeans(n_clusters = num_clusters, random_state = 428)
  clustering.fit(X)
  inertia_list.append(clustering.inertia_)

# plot the intertia curve
plt.plot(range(1,10), inertia_list)
plt.scatter(range(1,10), inertia_list)
plt.scatter(3,inertia_list[3],marker="X", s=300,c="r")
plt.xlabel("Number of Clusters", size=13)
plt.ylabel("inertia Value", size=13)
plt.title("Different Inertia Values for Different Number of Clusters")
```

As seen, WSS is greatly reduced when k increases from one to two. Another substantial reduction in WSS occurs at k = 3. However improvement in WSS is fairly linear for k>3. Therefore, the k-means analysis will be conducted for k=3
